```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load dataset
data = load_breast_cancer()

X = data.data
y = data.target

# Convert to DataFrame (for inspection)
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

df.head()
```

```python
df.isnull().sum()
```

✅ Explanation:

The dataset contains no missing values.

Therefore, no imputation is required.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

✅ Why Scaling is Necessary?

Features like mean radius, mean area, and mean smoothness have different ranges.

Algorithms like:

Logistic Regression

SVM

k-NN
are distance-based or gradient-based, so scaling prevents bias toward large magnitude features.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

1. Logistic Regression
🔹 How It Works:

Uses a sigmoid function to model probability.

Finds a linear decision boundary.

Optimizes using maximum likelihood.

🔹 Why Suitable?

Dataset is relatively clean and linearly separable.

Works well for binary classification.

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test, lr_pred)
```

2. Decision Tree Classifier
🔹 How It Works:

Splits data based on feature thresholds.

Uses criteria like Gini or Entropy.

Builds hierarchical decision rules.

🔹 Why Suitable?

Captures non-linear relationships.

Easy to interpret.

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

dt_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, dt_pred)
```

3. Random Forest Classifier
🔹 How It Works:

Ensemble of multiple decision trees.

Uses bagging (bootstrap sampling).

Reduces overfitting.

🔹 Why Suitable?

Handles feature interactions well.

More robust than a single tree.

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)
```

4. Support Vector Machine (SVM)
🔹 How It Works:

Finds optimal hyperplane maximizing margin.

Can use kernels for non-linear classification.

🔹 Why Suitable?

Works well in high-dimensional spaces.

Effective for small to medium datasets.

```python
svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)

svm_pred = svm.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)
```

5. k-Nearest Neighbors (k-NN)
🔹 How It Works:

Classifies based on majority vote of nearest neighbors.

Uses distance metrics (usually Euclidean).

🔹 Why Suitable?

Simple and effective.

Works well when classes are well separated.

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

knn_pred = knn.predict(X_test)
knn_acc = accuracy_score(y_test, knn_pred)
```

```python
results = pd.DataFrame({
    'Model': [
        'Logistic Regression',
        'Decision Tree',
        'Random Forest',
        'SVM',
        'k-NN'
    ],
    'Accuracy': [
        lr_acc,
        dt_acc,
        rf_acc,
        svm_acc,
        knn_acc
    ]
})

results.sort_values(by='Accuracy', ascending=False)
```
